```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'dataset\current-power.csv',delimiter=",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.Figure()
plt.scatter(x,y,
            color='red')
plt.title("Device Current vs Power Response")
plt.ylabel('Power (X)')
plt.xlabel('Current (Y)')
plt.show()
```

```python
df.corr().round(3)
```

```python
#regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#scatter point
plt.Figure(figsize=(7,4))
plt.scatter(x, y,
            color="#2A0053",
            alpha=0.6,
            label='data points')


plt.plot(x,line,
         color="#FF0000",
         label=f'regression line(r={r_value:.3f})')

plt.Figure()
plt.scatter(x,y,
            color="#000100")
plt.title("Device Current vs Power Response")
plt.ylabel('Power (X)')
plt.xlabel('Current (Y)')
plt.legend()
plt.show()




```

## <hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr><b>AIR QUALITY<hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr>

```python
air = pd.read_csv(r'dataset\air-quality.csv',delimiter=",")

air
```

## <hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr><b>NOx VS HUMIDITY<hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr>

```python
x = air['NOx']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

plt.figsize=(7,4)
plt.scatter(x, y,
            color="#2A0053",
            alpha=0.6,
            label='data points')
plt.plot(x,line,
         color="#FF0000",
            label=f'regression line(r={r_value:.3f})')
plt.title("NOx vs Relative Humidity")
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ppb)')
plt.legend()
plt.show()
```

## <hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr><b>NO2 VS HUMIDITY<hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr>

```python
x = air['NO2']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(7, 4))
plt.scatter(x,y,
            color="#2A0053",
            alpha=0.2,
            label = "Data Points")  
plt.plot(x, line,
            color='red',
            label=f'Best Fit Line(r = {r_value:.3f})')
plt.grid()
plt.xlabel('NO2 (ug/cu.m)')
plt.ylabel('Relative Humidity (%)')
plt.title('Linear Regression: NO2 vs Relative Humidity')
plt.legend()
```

## <b>TEMPERATURE VS HUMIDITY<hr style="border:2.5px solid rgba(255, 0, 0, 1)"></hr>

```python
x = air['Temperature']
y = air['Relative_Humidity']

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

plt.figure(figsize=(7, 4))
plt.scatter(x,y,
            color='#2A0053',
            alpha=0.2,
            label = "Data Points")  
plt.plot(x, line,
            color='red',
            label=f'Best Fit Line(r = {r_value:.3f})')
plt.grid()
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.title('Linear Regression:Temperature vs Relative Humidity')
plt.legend()
```
