```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.DataFrame([1,2,2,3,3,3,4,4,5],
                  columns=['Dataset'])

df
```

```python
df.describe().round(2)
```

```python
df['Z-Score'] = stats.zscore(df['Dataset'],ddof=1)
df.round(2)
```

```python
df['Z-Score'].mean()
```

```python
df = pd.read_csv(r"dataset\current-test.csv",
                 delimiter=',',
                 index_col=0)
df.head()
```

```python
df['Current'][2]
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Z-score'] = stats.zscore(df['Current'], ddof=1)
df
```

```python
df['Z-score'].describe().round(1)
```


## <hr style="border:2.5px solid rgba(157, 255, 0, 1)"><b> HISTOGRAM <hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr>

```python
plt.hist(df['Current'], bins=5,
         alpha=0.7,
         color="#FF0000",
         edgecolor='black')

plt.title('Current Distribution Histogram',
          fontsize=14,
          fontfamily='Sans-Serif',
          fontweight='bold')
plt.xlabel('Current (A)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```


## <hr style="border:2.5px solid rgba(157, 255, 0, 1)"><b> NORMAL DISTRIBUTION <hr style="border:2.5px solid rgba(157, 255, 0, 1)"></hr>

```python
mean=df['Current'].mean()
std=df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p, 'k', color='black', linewidth=2)
plt.title('Current Distribution Curve',
            fontsize=14,
            fontfamily='Sans-Serif',
            fontweight='bold')
plt.xlabel('Current (A)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

```python
plt.hist(df['Current'], bins=5,
         density=True,
         alpha=0.7,
         color="#FF0000",
         edgecolor='black')

plt.title('Current Distribution Histogram',
          fontsize=14,
          fontfamily='Sans-Serif',
          fontweight='bold')
plt.xlabel('Current (A)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

mean=df['Current'].mean()
std=df['Current'].std(ddof=1)

x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p, 'k', color='black', linewidth=2)
plt.title('Current Distribution Curve',
            fontsize=14,
            fontfamily='Sans-Serif',
            fontweight='bold')
plt.xlabel('Current (A)', fontsize=12)
plt.ylabel('Density', fontsize=12)
```

## <b>STANDARD NORMAL DISTRIBUTION (0,1)<hr style="border:2.5px solid rgba(255, 0, 0, 1)"></hr>

```python
x_min = df['Z-score'].min()
x_max = df['Z-score'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p, 'k', color='black', linewidth=2)
plt.figsize=(8,5)
plt.title('Standard Normal Distribution of Current Distribution Curve',
            fontsize=14,
            fontfamily='Sans-Serif',
            fontweight='bold')
plt.xlabel('Z-score', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.show()
```
