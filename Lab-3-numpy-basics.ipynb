# Measures of Central tendency

```python
# Jaguar vs Panther

resistance = np.array([
    [3,9.1,9.3,9.5,9.8,9.8,10.4,10.4,10.5,10.6,11],
    [9.1,9.1,9.1,9.2,9.2,9.9,10.1,10.8,11,20,np.nan]


])
```

```python
np.shape(resistance)
```

```python
res_mean = np.nanmean(resistance,axis=1)

```

```python
print("Jaguar mean:" + str((res_mean[0]).round(2))+ "ohms")
print ("Panther mean:" + str(res_mean[1]))
```

```python
print(f'Jaguar:{res_mean[0]:.2f} ohms')
```

```python
res_median = np.nanmedian(resistance,axis=1)
res_median
```

```python
res_mode = stats.mode(resistance,axis=1)
res_mode
```

```python
jag_res = resistance[0]

print("Jaguar mode:")
for i in multimode(jag_res):
    print(f'{i}: ohms')
```

```python
pan_res = resistance[1]

print("Panther mode:")
for l in multimode(pan_res):
    print(f'{l}: ohms')
```

# Measures of Central tendency

```python
# Jaguar vs Panther

resistance = np.array([
    [3,9.1,9.3,9.5,9.8,9.8,10.4,10.4,10.5,10.6,11],
    [9.1,9.1,9.1,9.2,9.2,9.9,10.1,10.8,11,20,np.nan]


])
```

```python
res_max = np.nanmax(resistance, axis=1)
res_min = np.nanmin(resistance, axis=1)

res_range = res_max - res_min
res_range
```

```python
print(f'Jaguar_range:{res_range[0]:.2f} ohms')
print(f'Panther_range:{res_range[1]:.2f} ohms')
```

```python
res_Q1 = np.nanpercentile(resistance,25,axis=1)
res_Q1
```

```python
print(f'Jaguar IQR: {res_Q1[0]:.2f} ohms')
```

```python
res_var = np.nanvar(resistance,ddof=1,axis=1)

print(f'Jaguar Variance:{res_var[0]:.2f} ohms')
```

```python
res_std = np.nanstd(resistance,ddof=1,axis=1)

print(f'Jaguar Variance:{res_var[0]:.2f} ohms')
print(f' Variance:{res_var[1]:.2f} ohms')
```
